```r
arg.list <- list(col.regions = group.colors(3),
                  at = seq(0, 2, 1), main = paste("Fire seasons' form"))
v <- numeric(nrow(df.log))
if (length(naind) == 0){
    v <- clus$classification
} else {
    v[naind] <- NA
    v[-naind] <- fireSeasonMedian$form[-naind]
}
df1 <- cbind.data.frame(coords, v)
coordinates(df1) <- c(1,2)
gridded(df1) <- TRUE
arg.list[["sp.layout"]] <- list("sp.lines", coast.lines)
arg.list[["obj"]] <- df1
arg.list[["zcol"]] <- 1
arg.list[["ylim"]] <- c(-90,90)
arg.list[["xlim"]] <- c(-180,180)
do.call("spplot", arg.list) %>% print()
```

```r
arg.list <- list(col.regions = group.colors(12), at = seq(0, 12, 1), main = "Unimodal fire seasons' start month")
v <- numeric(nrow(df.log))

v[-which(fireSeasonMedian$form == 1)] <- NA
v[which(fireSeasonMedian$form == 1)] <- fireSeasonMedian$start.1[which(fireSeasonMedian$form == 1)]

df1 <- cbind.data.frame(coords, v)
coordinates(df1) <- c(1,2)
gridded(df1) <- TRUE
arg.list[["sp.layout"]] <- list("sp.lines", coast.lines)
arg.list[["obj"]] <- df1
arg.list[["zcol"]] <- 1
arg.list[["ylim"]] <- c(-90,90)
arg.list[["xlim"]] <- c(-180,180)
do.call("spplot", arg.list) %>% print()
```

```r
arg.list <- list(col.regions = group.colors(8), at = seq(0, 8, 1), main = "Unimodal fire seasons' duration")
v <- numeric(nrow(df.log))

v[-which(fireSeasonMedian$form == 1)] <- NA

v[which(fireSeasonMedian$form == 1)] <- fireSeasonMedian$end.1[which(fireSeasonMedian$form == 1)] - fireSeasonMedian$start.1[which(fireSeasonMedian$form == 1)]
v[which(fireSeasonMedian$form == 1)][which(v[which(fireSeasonMedian$form == 1)]<0)] = 12 - abs(v[which(fireSeasonMedian$form == 1)][which(v[which(fireSeasonMedian$form == 1)]<0)])
v[which(fireSeasonMedian$form == 1)] = v[which(fireSeasonMedian$form == 1)] + 1

df1 <- cbind.data.frame(coords, v)
coordinates(df1) <- c(1,2)
gridded(df1) <- TRUE
arg.list[["sp.layout"]] <- list("sp.lines", coast.lines)
arg.list[["obj"]] <- df1
arg.list[["zcol"]] <- 1
arg.list[["ylim"]] <- c(-90,90)
arg.list[["xlim"]] <- c(-180,180)
do.call("spplot", arg.list) %>% print()
```

```r
arg.list <- list(col.regions = group.colors(8), at = seq(0, 8, 1), main = "Total length of whole fire seasons")
v <- numeric(nrow(df.log))

v[-which(fireSeasonMedian$form == 1)] <- NA

v[which(fireSeasonMedian$form == 1)] <- fireSeasonMedian$end.1[which(fireSeasonMedian$form == 1)] - fireSeasonMedian$start.1[which(fireSeasonMedian$form == 1)]
v[which(fireSeasonMedian$form == 1)][which(v[which(fireSeasonMedian$form == 1)]<0)] = 12 - abs(v[which(fireSeasonMedian$form == 1)][which(v[which(fireSeasonMedian$form == 1)]<0)])
v[which(fireSeasonMedian$form == 1)] = v[which(fireSeasonMedian$form == 1)] + 1

v[which(fireSeasonMedian$form == 2)] <- fireSeasonMedian$end.1[which(fireSeasonMedian$form == 2)] - fireSeasonMedian$start.1[which(fireSeasonMedian$form == 2)]
v[which(fireSeasonMedian$form == 2)][which(v[which(fireSeasonMedian$form == 2)]<0)] = 12 - abs(v[which(fireSeasonMedian$form == 2)][which(v[which(fireSeasonMedian$form == 2)]<0)])
v[which(fireSeasonMedian$form == 2)] = v[which(fireSeasonMedian$form == 2)] + 1

v[which(fireSeasonMedian$form == 2)] <- v[which(fireSeasonMedian$form == 2)] + fireSeasonMedian$end.2[which(fireSeasonMedian$form == 2)] - fireSeasonMedian$start.2[which(fireSeasonMedian$form == 2)] + 1

df1 <- cbind.data.frame(coords, v)
coordinates(df1) <- c(1,2)
gridded(df1) <- TRUE
arg.list[["sp.layout"]] <- list("sp.lines", coast.lines)
arg.list[["obj"]] <- df1
arg.list[["zcol"]] <- 1
arg.list[["ylim"]] <- c(-90,90)
arg.list[["xlim"]] <- c(-180,180)
do.call("spplot", arg.list) %>% print()
```
