# Clustering with the 5% restriction

This notebook contains the new clustering and fire seasons which have been performing using the new data. Also there are some plots summarizing the fire seasons. Now, each cluster must have more than 5% of the points of the biome, without taking into account the NA values. As a result, the number of clusters of each biome must be less or equal than 20.

Fire seasons were calculated using the 75 percentile.

The plots of the form of the fire seasons show the 25 percentile as a green line, the median as a red one and the 75 percentile with a blue line.

```r
#jpeg(file="Figures/fs_start_month.jpeg")

par(mfrow = c(3,1))
for (i in 1:13){
    arg.list <- list(col.regions = group.colors(12), at = seq(0, 12, 1), main = paste("Start month biome", toString(i)))
    v <- numeric(nrow(df.log))

    v[-which(fireSeason_def_2$form == 1)] <- NA
    v[which(fireSeason_def_2$form == 1)] <- fireSeason_def_2$start.1[which(fireSeason_def_2$form == 1)]
    v[-which(fireSeason_def_2$BIOME == i)] <- NA

    df1 <- cbind.data.frame(masked_coords, v)
    coordinates(df1) <- c(1,2)
    gridded(df1) <- TRUE
    arg.list[["sp.layout"]] <- list("sp.lines", coast.lines)
    arg.list[["obj"]] <- df1
    arg.list[["zcol"]] <- 1
    arg.list[["ylim"]] <- c(-90,90)
    arg.list[["xlim"]] <- c(-180,180)
    do.call("spplot", arg.list) %>% print()
    
    
    arg.list <- list(col.regions = group.colors(12), at = seq(0, 12, 1), main = paste("Fire season duration (biome ", toString(i), ')', sep = ''))
    v <- numeric(nrow(df.log))

    v[-which(fireSeason_def_2$form == 1)] <- NA

    v[which(fireSeason_def_2$form == 1)] <- fireSeason_def_2$end.1[which(fireSeason_def_2$form == 1)] - fireSeason_def_2$start.1[which(fireSeason_def_2$form == 1)]
    v[which(fireSeason_def_2$form == 1)][which(v[which(fireSeason_def_2$form == 1)]<0)] = 12 - abs(v[which(fireSeason_def_2$form == 1)][which(v[which(fireSeason_def_2$form == 1)]<0)])
    v[which(fireSeason_def_2$form == 1)] = v[which(fireSeason_def_2$form == 1)] + 1
    v[-which(fireSeason_def_2$BIOME == i)] <- NA

    df1 <- cbind.data.frame(masked_coords, v)
    coordinates(df1) <- c(1,2)
    gridded(df1) <- TRUE
    arg.list[["sp.layout"]] <- list("sp.lines", coast.lines)
    arg.list[["obj"]] <- df1
    arg.list[["zcol"]] <- 1
    arg.list[["ylim"]] <- c(-90,90)
    arg.list[["xlim"]] <- c(-180,180)
    do.call("spplot", arg.list) %>% print()
}

#dev.off()
```
